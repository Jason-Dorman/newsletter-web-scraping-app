```python
from splinter import Browser
from bs4 import BeautifulSoup
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://www.sandiegouniontribune.com/news/politics'
browser.visit(url)
```

```python
# UT stories first group
html = browser.html
soup = BeautifulSoup(html, 'html.parser')

stories = soup.find_all(class_='PromoMedium-content')

for story in stories:
    title = story.find(class_='PromoMedium-title')
    summary = story.find(class_='PromoMedium-description').text
    link = title.a['href']
    print('UT Stories')
    print('------------------------')
    print('Title: ', title.text)
    print('Summary: ', summary)
    print('')
    print('Link: ', link)
    print('')
```

```python
# UT trending stories

t_stories = soup.find_all(class_='PromoXSmall-title')

for t_story in t_stories:
    link = t_story.a['href']
    print('UT Trending Stories:')
    print('------------------------------------------------------------------------')
    print(t_story.text)
    print('link: ', link)
    print('')
```

```python
# voice
v_url = 'https://www.voiceofsandiego.org/'
browser.visit(v_url)
```

```python
v_html = browser.html
v_soup = BeautifulSoup(v_html, 'html.parser')

v_title = v_soup.find('h1', class_='excerpt-title').text
v_summary = v_soup.find(class_='excerpt-lede').text
v_story = v_soup.find('h1', class_='excerpt-title')
v_link = v_story.a['href']

v_titles = v_soup.find_all(class_='thumb-align')

print('Voice Story 1 Title: ', v_title)
print('Summary: ', v_summary)
print('Link: ', v_link)
print('')
print('---------------------------------------')

for v_stories in v_titles:
    v_links_2 = v_stories.a['href']
    print('Voice Story: ', v_stories.text)
    print('Link: ', v_links_2)
    print('')
    print('-----------------------------------')
```

```python
# SD Reader
sd_url = 'https://www.sandiegoreader.com/news/politics/'
browser.visit(sd_url)
```

```python
sd_html = browser.html
sd_soup = BeautifulSoup(sd_html, 'html.parser')

results = sd_soup.find_all(class_='item')
link_starter = 'https://www.sandiegoreader.com'

for result in results:
    title = result.find('h3').text
    summary = result.find_all('p')
    link = result.a['href']
    print('SD Reader Story')
    print('-----------------------')
    print('Title: ', title)
    print('Summary: ', summary)
    print('Link: ', link_starter + link)
    print('')
```

```python
# Times of San Diego
times_url = 'https://timesofsandiego.com/category/politics/'
browser.visit(times_url)
```

```python
times_html = browser.html
times_soup = BeautifulSoup(times_html, 'html.parser')

times_stories = times_soup.find_all('article')

for times_story in times_stories:
    title = times_story.find('h2').text
    summary = times_story.find(class_='entry-summary').text
    link = times_story.a['href']
    print('Times of San Diego Story')
    print('---------------------------------')
    print('Title: ', title)
    print('Summary: ', summary)
    print('Link: ', link)
    print('')
```
